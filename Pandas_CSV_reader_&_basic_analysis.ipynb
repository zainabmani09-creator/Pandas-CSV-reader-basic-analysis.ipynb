```python
import pandas as pd
import numpy as np
import json

with open('titanic-machine-learning-from-disaster-metadata.json', 'r') as f:
    data = json.load(f)

df = pd.json_normalize(data)
```

```python
df['description'].fillna('Not Available', inplace=True)
```

```python
df['version'] = df['version'].interpolate()
```

```python
df.dropna(subset=['name'], inplace=True)
```

```python
# Value counts for dataset name
df['name'].value_counts()
```

```python
# Group by creator name and count datasets
df.groupby('creator.name').size()
```

```python
pivot = df.pivot_table(
    index='publisher.name',
    values='version',
    aggfunc='count'
)

pivot
```

```python
df['version_cumsum'] = df['version'].cumsum()
```

```python
files = df[['distribution']].explode('distribution').copy()
files['contentSize'] = files['distribution'].apply(
    lambda x: float(x.get('contentSize').split()[0]) if isinstance(x.get('contentSize'), str) else 0.0
)

files['rolling_3_size'] = files['contentSize'].rolling(3).sum()
```

```python
files['size_rank'] = files['contentSize'].rank(ascending=False)
```

```python
files['previous_file_size'] = files['contentSize'].shift(1)
```

```python
files[['contentSize', 'rolling_3_size', 'size_rank', 'previous_file_size']].head()
```
